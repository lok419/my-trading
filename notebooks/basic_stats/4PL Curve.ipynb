```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def four_pl_curve(x,A,D,IC50,n,m):    
    return A + (D-A)/(1+(x/IC50)**n)**m
```

```python
x = np.arange(0,1,0.01)
```

```python
A = 15
D = 0.25
IC50 = 0.5
n = 5
m = 1
```

```python
plt.plot(x,four_pl_curve(x,A,D,IC50,3,2))
plt.plot(x,four_pl_curve(x,A,D,IC50,4,2))
plt.plot(x,four_pl_curve(x,A,D,IC50,5,2))
plt.plot(x,four_pl_curve(x,A,D,IC50,6,2))
```

```python
plt.plot(x,four_pl_curve(x,A,D,IC50,n,1))
plt.plot(x,four_pl_curve(x,A,D,IC50,n,2))
plt.plot(x,four_pl_curve(x,A,D,IC50,n,3))
plt.plot(x,four_pl_curve(x,A,D,IC50,n,4))
```

```python
plt.plot(x,four_pl_curve(x,A,D,IC50,3,m))
plt.plot(x,four_pl_curve(x,A,D,IC50,4,m))
plt.plot(x,four_pl_curve(x,A,D,IC50,5,m))
plt.plot(x,four_pl_curve(x,A,D,IC50,6,m))
```
