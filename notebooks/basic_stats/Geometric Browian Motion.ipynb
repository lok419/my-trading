## Geometric Browian Motion

Differentiate Equation

$$dX_t = \mu X_tdt+\sigma_tX_tdW_t$$

Apply Ito's Lemma to $LogX_t$, then we can derive the Closed Form Solution

$$X_t = X_0exp\left((u-\frac{\sigma^2}{2})t + \sigma W_t \right)$$

where $W_t \sim N(0,t)$

## Simluation by Monte Carlo

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
M = 1000       # number of simuation
mu = 0.03      # drift rate
sigma = 0.6    # volatiliy
T = 600        # time steps
X0 = 200.0     # initial values

dt = 0.001
X = np.full(shape=(M,T), fill_value=X0)
W = np.random.normal(0,np.sqrt(dt),size=(M,T))

# for i in range(M):            
#     X_ = np.exp((mu-0.5*sigma**2)*dt + sigma * np.random.normal(0,np.sqrt(dt), size=(T)))        
#     X[i] = X0 * X_.cumprod()        

for i in range(1,T):
    X[:,i] = X[:,i-1] * np.exp((mu-0.5*sigma**2)*dt + sigma * W[:,i])        
```

```python
for ys in X[:10]:
    plt.plot(ys)
plt.grid()
plt.title("Simulated GBM paths")
plt.show()

plt.hist(X[:,-1], bins=30)
plt.title("Distribution at last step")
plt.grid()
plt.show()
```
