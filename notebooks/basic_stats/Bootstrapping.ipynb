```python
import numpy as np
import matplotlib.pyplot as plt
```

# True Population

```python
mean = 5
std = 3
n = 10000
x = np.random.normal(mean,std,n)

plt.figure(figsize=(4,4))
plt.hist(x,bins=n//100)
plt.axvline(mean,color="red",linestyle="--")
plt.show()
```

# Bootstrapped

```python
k = 5000
bootstrapped_n = 10

x_b = []

for _ in range(k):
    x_b.append(np.mean(np.random.normal(mean,std,bootstrapped_n)))
    
plt.figure(figsize=(7,4))
plt.hist(x,bins=n//100,alpha=0.5,label="true population")
plt.hist(x_b,bins=k//100, alpha=0.5, label="bootstrapped mean")
plt.axvline(mean,color="red",linestyle="--")
plt.legend()
plt.show()

print(f'standard error of mean: {np.std(x_b)}')
print(f'sample standard derivation: {np.std(x_b) * np.sqrt(bootstrapped_n)}')
print(f'population standard derivation: {std}')
```
